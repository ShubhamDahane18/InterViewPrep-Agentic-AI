```python
import whisper
import sounddevice as sd
import numpy as np
import tempfile
import queue
import threading
import os

# Load Whisper model (choose "tiny", "base", "small", "medium", or "large")
model = whisper.load_model("base")

# Audio settings
samplerate = 16000  # Whisper works best at 16kHz
blocksize = 4000    # Buffer size
q = queue.Queue()

# Callback to collect audio chunks
def audio_callback(indata, frames, time, status):
    if status:
        print(status, flush=True)
    q.put(indata.copy())

# Function to record & transcribe in chunks
def transcribe_stream():
    with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as f:
        while True:
            data = q.get()
            if data is None:
                break
            # Save chunk temporarily
            np.save(f.name + ".npy", data)
            audio_data = np.load(f.name + ".npy")
            audio_data = audio_data.flatten()

            # Transcribe
            result = model.transcribe(audio_data, fp16=False)
            print("You said:", result["text"])

# Start recording
print("ğŸ¤ Listening... Press Ctrl+C to stop.")
stream = sd.InputStream(callback=audio_callback, channels=1, samplerate=samplerate, blocksize=blocksize)
with stream:
    threading.Thread(target=transcribe_stream, daemon=True).start()
    try:
        while True:
            pass
    except KeyboardInterrupt:
        print("\nStopped.")
        q.put(None)
```

```python
from google import genai
from google.genai import types
import wave

# Set up the wave file to save the output:
def wave_file(filename, pcm, channels=1, rate=24000, sample_width=2):
   with wave.open(filename, "wb") as wf:
      wf.setnchannels(channels)
      wf.setsampwidth(sample_width)
      wf.setframerate(rate)
      wf.writeframes(pcm)

client = genai.Client()

response = client.models.generate_content(
   model="gemini-2.5-flash-preview-tts",
   contents="Say cheerfully: Have a wonderful day!",
   config=types.GenerateContentConfig(
      response_modalities=["AUDIO"],
      speech_config=types.SpeechConfig(
         voice_config=types.VoiceConfig(
            prebuilt_voice_config=types.PrebuiltVoiceConfig(
               voice_name='Kore',
            )
         )
      ),
   )
)

data = response.candidates[0].content.parts[0].inline_data.data

file_name='out.wav'
wave_file(file_name, data) # Saves the file to current directory
```
